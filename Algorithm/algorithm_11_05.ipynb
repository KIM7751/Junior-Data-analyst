# 오늘의 지식
### split( )
- 함수의 모양

  - 문자열.split()

  - 문자열.split('구분자')

  - 문자열.split('구분자', 분할횟수)

  - 문자열.split(sep='구분자', maxsplit=분할횟수)

- 문자열을 일정한 규칙으로 잘라서 리스트로 만들어주는 함수

- 문자열.split(sep, maxsplit)함수는 문자열을 maxsplit 횟수만큼 sep의 구분자를 기준으로 문자열을 구분하여 "리스트"로 만듦

  - sep 파라미터 : 기본값 none으로 띄어쓰기, 엔터를 구분자로 해 문자열을 자름
  - maxsplit 파라미터 : 기본값 -1, 제한없이 자를 수 있을 때까지 문자열 전체를 자름
  - maxsplit 를 생략이 가능하지만 앞에 sep 파라미터가 있어야지만 가능함
    - 문자열.split(maxsplit=1), 문자열을 1번만 자르게 됨

## range( )
- range(start, stop, step)
  - range(stop) - 마지막 숫자는 포함하지 않음
  - range(start, stop) - 첫번째 인자는 시작하는 숫자가 됨
  - range(start, stop, step) - 마지막 인자 step은 숫자의 간격을 나타냄, step으로 음수를 지정 가능
- range() 함수의 결과는 반복가능(iterable)하기 때문에 for문을 사용해 출력할 수 있음
- ※ for문에서 range를 사용하지 않고 int를 직접 사용하는 것은 불가능 ※

# 알고리즘 문제 풀기
