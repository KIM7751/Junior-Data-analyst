# 오늘의 지식



## (1 ≤ m ≤ n ≤ 101000, m과 n은 10진수 정수)
- 내가 푸는 알고리즘 문제는 채점시 해당 범위에 있는 숫자로 진행함

## int
- 파이썬에서 int 값 출력을 할 수 있는 범위가 2**16임

# 알고리즘 문제 풀기


```python
# https://www.acmicpc.net/problem/10998

a, b = map(int, input().split())
print(a * b)
```

- 소수점이 나오면 안되니깐 / 가 아닌 //
- 런타임오류(overflow)난 이유는 파이썬에서 int 값 출력을 할 수 있는 범위가 2**16임 

```python
# https://www.acmicpc.net/problem/1271
# 가진돈 = n, 생명체 = m
n, m = map(int, input().split())
print(n // m)
print(n % m)
```
