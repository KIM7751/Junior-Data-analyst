```python
# 라이브러리 적용
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 로드
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(path)
iris.head()
```

```python
# 원핫인코딩
iris = pd.get_dummies(iris)
```

```python
# 종속변수, 독립변수 분리
X = iris.drop(['품종_setosa', '품종_versicolor', '품종_virginica'], axis=1)
y = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
```

```python
print(X.shape, y.shape)
```

```python
# 모델 구조 만들기
Input = tf.keras.layers.Input(shape=[4])
Output = tf.keras.layers.Dense(3, activation='softmax')(Input) # 종속변수 3개, 분류 모델 활성함수 지정 
model = tf.keras.models.Model(Input, Output)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

- 분류 모델이 분류를 추측하는 것도 확률로 표현함
- 이렇게 만들어주는 활성화 함수가 softmax, sigmoid 임
  - softmax -> multi-classification, 확률의 총합 = 1
  - sigmoid -> binary-classification




```python
# 모델 학습
model.fit(X, y, epochs=1000, verbose=0)
model.fit(X, y, epochs=10)
```

```python
# 모델 이용
print(model.predict(X[:5]))
print(y[:5])
```

```python
print(model.predict(X[-5:]))
print(y[-5:])
```
