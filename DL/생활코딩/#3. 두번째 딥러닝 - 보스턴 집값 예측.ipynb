```python
# 라이브러리
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 불러오기 
url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(url)
boston
```

```python
# 독립변수, 종속변수 분리
X = boston.drop(['medv'], axis=1)
y = boston['medv']
```

```python
print(X.shape, y.shape)
```

```python
# 모델 구조 생성
Input = tf.keras.layers.Input(shape=[13]) # 초기 삽입 값 개수 지정(독립변수 개수)
Output = tf.keras.layers.Dense(1)(Input) # 출력값 지정(종속변수 개수, 수식 모양 구성단계)
model = tf.keras.Model(Input, Output)
model.compile(loss='mse')
```

```python
# 모델 학습
model.fit(X, y, epochs=1000, verbose=0)
model.fit(X, y, epochs=20)
```

```python
# 예측
print(model.predict(X[5:10]))
print(y[5:10])
```

```python
# 모델 수식
print(model.get_weights())
```
