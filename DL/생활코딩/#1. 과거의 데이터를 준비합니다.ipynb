```python
import pandas as pd
```

```python
# 데이터 불러오기
lemon_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
boston_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
iris_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
lemon = pd.read_csv(lemon_path)
boston = pd.read_csv(boston_path)
iris = pd.read_csv(iris_path)
```

```python
# 데이터 모양으로 확인하기
print(lemon.shape)
print(boston.shape)
print(iris.shape)
```

```python
# 칼럼 이름 출력
print(lemon.columns)
print(boston.columns)
print(iris.columns)
```

```python
x = lemon[['온도']]  # 대괄호 하나 사용시 "series" 두개 사용시 "dataframe"
y = lemon[['판매량']]
print(x.shape, y.shape)
```

```python
x = boston.drop(['medv'], axis=1) 
y = boston[['medv']]
print(x.shape, y.shape)
```

```python
x = iris.drop(['품종'], axis=1)  
y = iris[['품종']]
print(x.shape, y.shape)
```
