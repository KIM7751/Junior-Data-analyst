```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

- 기존에는 퍼셉트론, 신경망 하나로 모델을 만들었음 
- 히든레이어를 넣어서 우리가 흔히 말하는 딥러닝 실습을 진행

# 보스턴 집값 예측 

```python
# 데이터 로드
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(path)
boston
```

```python
# 종속변수, 독립변수
X = boston.drop(['medv'], axis=1)
y = boston[['medv']]
```

```python
# 모델의 구조 생성
Input = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(10, activation='swish')(Input) # 노드 10개 생성, 활성화 함수 "스위시"
Output = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(Input, Output)
model.compile(loss='mse')
```

```python
# 확인
model.summary() # param 가중치 w, b 개수
```

```python
# 학습
model.fit(X, y, epochs=1000, verbose=0)
model.fit(X, y, epochs=10)
```

```python
# 예측
print(model.predict(X[:5]))
print(y[:5])
```

# 아이리스 품종 분류

```python
# 데이터 로드
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(path)
```

```python
# 원핫 인코딩
iris = pd.get_dummies(iris)
iris
```

```python
# 종속변수, 독립변수
X = iris.drop(['품종_setosa', '품종_versicolor', '품종_virginica'], axis=1)
y = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
```

```python
# 모델의 구조 생성
Input = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(Input) # 노드 8개 갖는 히든레이어 3개 생성 
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Output = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(Input, Output)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```python
# 확인
model.summary()
```

```python
# 학습
model.fit(X, y, epochs=10000, verbose=0)
model.fit(X, y, epochs=10)
```

```python
print(model.predict(X[:5]))
print(y[:5])
```
